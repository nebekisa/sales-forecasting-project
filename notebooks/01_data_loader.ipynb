```python
# Cell 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries imported successfully!")
print(f"Pandas version: {pd.__version__}")
```

```python
# Cell 2: Define paths to your data
# IMPORTANT: Run this cell FIRST before Cell 3

train_path = Path('../data/raw/train.csv')
test_path = Path('../data/raw/test.csv')
store_path = Path('../data/raw/store.csv')

print(f"Current working directory: {Path.cwd()}")
print(f"\nChecking files:")
print(f"Train file exists: {train_path.exists()}")
print(f"Test file exists: {test_path.exists()}")
print(f"Store file exists: {store_path.exists()}")
```

```python
# Cell 3: Load all three CSV files
print("Loading data...")

# Load training data
train_df = pd.read_csv(train_path)
print(f"\n✓ Train data loaded: {train_df.shape[0]:,} rows, {train_df.shape[1]} columns")

# Load test data
test_df = pd.read_csv(test_path)
print(f"✓ Test data loaded: {test_df.shape[0]:,} rows, {test_df.shape[1]} columns")

# Load store data
store_df = pd.read_csv(store_path)
print(f"✓ Store data loaded: {store_df.shape[0]:,} rows, {store_df.shape[1]} columns")
```

```python
# Cell 4: Look at training data
print("="*60)
print("TRAINING DATA - First 5 rows")
print("="*60)
train_df.head()
```

```python
# Cell 5: Training data information
print("="*60)
print("TRAINING DATA - Info")
print("="*60)
train_df.info()
```

```python
# Cell 6: Statistical summary
print("="*60)
print("TRAINING DATA - Statistics")
print("="*60)
train_df.describe()
```

```python
# Cell 7: Missing values in training data
print("="*60)
print("TRAINING DATA - Missing Values")
print("="*60)

missing_train = train_df.isnull().sum()
missing_percent = (missing_train / len(train_df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_train,
    'Percentage': missing_percent
})

missing_df[missing_df['Missing Count'] > 0]
```

```python
# Cell 8: Look at store data
print("="*60)
print("STORE DATA - First 5 rows")
print("="*60)
store_df.head()
```

```python
# Cell 9: Store data information
print("="*60)
print("STORE DATA - Info")
print("="*60)
store_df.info()
```

```python
# Cell 10: Missing values in store data
print("="*60)
print("STORE DATA - Missing Values")
print("="*60)

missing_store = store_df.isnull().sum()
missing_percent_store = (missing_store / len(store_df)) * 100

missing_store_df = pd.DataFrame({
    'Missing Count': missing_store,
    'Percentage': missing_percent_store
})

missing_store_df[missing_store_df['Missing Count'] > 0]
```

```python
# Cell 11: Merge training data with store information
print("="*60)
print("MERGING TRAIN AND STORE DATA")
print("="*60)

# Merge on 'Store' column
merged_df = train_df.merge(store_df, on='Store', how='left')

print(f"Original train shape: {train_df.shape}")
print(f"Merged shape: {merged_df.shape}")
print(f"\nNew columns added from store data: {store_df.columns.tolist()}")
```

```python
# Cell 12: Look at merged data
merged_df.head()
```

```python
# Cell 13: Check the Date column
print("="*60)
print("DATE COLUMN ANALYSIS")
print("="*60)

print(f"Date column type: {merged_df['Date'].dtype}")
print(f"Date range: {merged_df['Date'].min()} to {merged_df['Date'].max()}")
print(f"Unique dates: {merged_df['Date'].nunique()}")
```

```python
# Cell 14: Save merged data for later use
output_path = Path('../data/processed/rossmann_merged.csv')
output_path.parent.mkdir(parents=True, exist_ok=True)

merged_df.to_csv(output_path, index=False)
print(f"✓ Merged data saved to: {output_path}")
print(f"  File size: {output_path.stat().st_size / (1024*1024):.2f} MB")
```

```python
# Cell 15: Summary of what we have
print("="*60)
print("DATA LOADING SUMMARY")
print("="*60)

print(f"\n📊 DATASETS LOADED:")
print(f"  • Train data: {train_df.shape[0]:,} rows, {train_df.shape[1]} columns")
print(f"  • Test data: {test_df.shape[0]:,} rows, {test_df.shape[1]} columns")
print(f"  • Store data: {store_df.shape[0]:,} rows, {store_df.shape[1]} columns")

print(f"\n📊 MERGED DATA:")
print(f"  • Shape: {merged_df.shape[0]:,} rows, {merged_df.shape[1]} columns")

print(f"\n🎯 TARGET VARIABLE:")
print(f"  • Sales: We will predict this")
print(f"  • Sales range: {merged_df['Sales'].min():,.0f} to {merged_df['Sales'].max():,.0f}")
print(f"  • Average sales: {merged_df['Sales'].mean():,.0f}")

print(f"\n✅ Data loading complete! Ready for cleaning.")
```
